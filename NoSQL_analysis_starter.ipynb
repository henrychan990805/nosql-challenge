## Notebook Set Up

```python
from pymongo import MongoClient
import pandas as pd
from pprint import pprint
```

```python
# Create an instance of MongoClient
mongo = MongoClient(port=27017)
```

```python
# assign the uk_food database to a variable name
db = mongo['uk_food']
```

```python
# review the collections in our database
print(db.list_collection_names())
```

```python
# assign the collection to a variable
establishments = db['establishments']
```

### 1. Which establishments have a hygiene score equal to 20?

```python
# Find the establishments with a hygiene score of 20
query = {'scores.Hygiene': 20}
hygiene_20 = establishments.find(query)


# Use count_documents to display the number of documents in the result
count_hygiene_20 = establishments.count_documents(query)
print(count_hygiene_20)
# Display the first document in the results using pprint
pprint(hygiene_20[0])
```

```python
# Convert the result to a Pandas DataFrame
hygiene_20_df = pd.DataFrame(list(hygiene_20))
# Display the number of rows in the DataFrame
print(hygiene_20_df['_id'].count())
# Display the first 10 rows of the DataFrame
hygiene_20_df.head()
```

### 2. Which establishments in London have a `RatingValue` greater than or equal to 4?

```python
# Find the establishments with London as the Local Authority and has a RatingValue greater than or equal to 4.
query = {"LocalAuthorityName":"City of London Corporation", "RatingValue" : {'$gte': 4}}
London_4_est = establishments.find(query)
# Use count_documents to display the number of documents in the result
London_4_est_count = establishments.count_documents(query)
print(London_4_est_count)
# Display the first document in the results using pprint
pprint(London_4_est[0])
```

```python
# Convert the result to a Pandas DataFrame
London_4_est_df = pd.DataFrame(list(London_4_est))
# Display the number of rows in the DataFrame
print(len(London_4_est_df))
# Display the first 10 rows of the DataFrame
London_4_est_df.head(10)
```

### 3. What are the top 5 establishments with a `RatingValue` rating value of 5, sorted by lowest hygiene score, nearest to the new restaurant added, "Penang Flavours"?

```python
# Search within 0.01 degree on either side of the latitude and longitude.
# Rating value must equal 5
# Sort by hygiene score
degree_search = 0.01
latitude = establishments.find({"BusinessName":"Penang Flavours"})[0]["geocode"]['latitude']
longitude = establishments.find({"BusinessName":"Penang Flavours"})[0]["geocode"]["longitude"]
query = {'$match': {"geocode.latitude": {'$lte': latitude+degree_search}, "geocode.latitude": {'$gte': latitude-degree_search},
         "geocode.longitude": {'$lte':longitude+degree_search}, "geocode.longitude": {'$gte': longitude-degree_search},
         "RatingValue": 5}}
sort = {"$sort": {'scores.Hygiene': 1}}
limit = {"$limit": 5}
pipeline = [query,sort,limit]
# Print the results
top_5 = list(establishments.aggregate(pipeline))
pprint(top_5)
```

```python
# Convert result to Pandas DataFrame
top_5_df = pd.DataFrame(top_5)
```

```python
top_5_df.head()
```

### 4. How many establishments in each Local Authority area have a hygiene score of 0?

```python
# Create a pipeline that:
# 1. Matches establishments with a hygiene score of 0
# 2. Groups the matches by Local Authority
# 3. Sorts the matches from highest to lowest
pipe = [
    {"$match": {"scores.Hygiene": 0}},
    {"$group": {"_id": "$LocalAuthorityName", "Count": {"$sum": 1} }},
    {"$sort": {"Count": -1}}]
low_hygiene = list(establishments.aggregate(pipe))
# Print the number of documents in the result
print(len(low_hygiene))
# Print the first 10 results
pprint(low_hygiene[0:10])
```

```python
# Convert the result to a Pandas DataFrame
low_hygiene_df = pd.DataFrame(low_hygiene)
# Display the number of rows in the DataFrame
print(len(low_hygiene_df))
# Display the first 10 rows of the DataFrame
low_hygiene_df.head(10)
low_hygiene_df
```
